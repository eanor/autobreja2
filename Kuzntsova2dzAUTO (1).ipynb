```python
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
```

```python
tags = []

for word in f_tokens:
    tags.append(str(morph.parse(word)[0].tag.POS))

tags
```

```python
from natasha import (
    Segmenter,
    MorphVocab,

    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    NewsNERTagger,

    PER,
    NamesExtractor,

    Doc
)



segmenter = Segmenter()
morph_vocab = MorphVocab()

emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)
ner_tagger = NewsNERTagger(emb)

names_extractor = NamesExtractor(morph_vocab)

natasha_dict = {}

for smth in f_tokens:
  doc = Doc(smth)
  doc.segment(segmenter)
  doc.tag_morph(morph_tagger)
  i = doc.sents[0].morph
  natasha_dict.update({i.tokens[0].text: i.tokens[0].pos})

natasha_dict
```

```python
def tagging(smth):
  if smth == 'ADJF':
    smth = 'ADJ'
  elif smth == 'ADJS':
    smth = 'ADJ'
  elif smth == 'ADVB':
    smth = 'ADV'
  elif smth == 'GRND':
    smth = 'V'
  elif smth == 'INFN':
    smth = 'V'
  elif smth == 'NOUN':
    smth = 'N'
  elif smth == 'NPRO':
    smth = 'PRON'
  elif smth == 'NUMR':
    smth = 'NUM'
  elif smth == 'PREP':
    smth = 'PR'
  elif smth == 'PRTF':
    smth = 'V'
  elif smth == 'PRTS':
    smth = 'V'
  elif smth == 'ADP':
    smth = 'PR'
  elif smth == 'DET':
    smth = 'PRON'
  elif smth == 'PART':
    smth = 'PRCL'
  elif smth == 'SCONJ':
    smth = 'CONJ'
  elif smth == 'AUX':
    smth = 'V'
  elif smth == 'CCONJ':
    smth = 'CONJ'
  else:
    smth =smth
  return smth
```

```python
py_morphy_list = []
for i in py_morphy_dict.values():
  py_morphy_list.append(tagging(i))
```
